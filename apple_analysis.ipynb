```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
apple = yf.download("AAPL", start="2023-01-01")
```

```python
print(apple.head())
```

```python
print(apple.columns)
```

```python
print(apple["Close"])
```

```python
apple["Close"].plot(figsize=(10,5))

plt.title("Apple Stock Price")
plt.xlabel("Date")
plt.ylabel("Price")

plt.show()
```

Overall upward trend, sharp decline in May 25 followed by a quick recovery. Early 25 shows highest period of volatility.

```python
apple["Return"] = apple["Close"].pct_change()
```

```python
apple["Return"].plot(figsize=(10,5))

plt.title("Apple Daily Returns")

plt.show()
```

```python
print("Average return:")
print(apple["Return"].mean())

print("Volatility:")
print(apple["Return"].std())
```
